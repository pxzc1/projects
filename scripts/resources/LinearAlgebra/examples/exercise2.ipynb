1. What's the transpose of this vector?

$$
\begin{bmatrix}
25 \\\\
2 \\\\
-3 \\\\
-23
\end{bmatrix}
$$

```python
import numpy as np

x = np.array([25,2,-3,-23])
print(x.T)
```

This transposition will turn this column vector into row vector, $(,1) \to (1,)$

2. Using algebriac notation, what are the dimensions of this matrix?

$$ y =
\begin{bmatrix}
42 & 4 & 7 & 99 \\\\
-99 & -3 & 17 & 22
\end{bmatrix}
$$

```python
import numpy as np

y = np.array([[42,4,7,99], [-99,-3,17,22]])
print(y.shape)
```

The dimension of this matrix $\mathbf{y}$ will be $(2,4)$. Because it has 2 rows and 4 columns.

3. Using algebriac notation, what is the position of the element in this matrix $\mathbf{y}$ with the value of 17?

```python
import numpy as np

y = np.array([[42,4,7,99], [-99,-3,17,22]])
print(y[1][2])
```

The element with value of 17 will be at the **second row, third element**. which in python will be `[1][2]`. (as python indexing start at 0)
