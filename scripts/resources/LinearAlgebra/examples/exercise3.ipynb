2. What is the Hadamard product of these matrices?

$$
\begin{bmatrix}
25 & 10 \\\\
-2 & 1
\end{bmatrix}
\odot
\begin{bmatrix}
-1 & 7 \\\\
10 & 8
\end{bmatrix}
$$

```python
import numpy as np

x = np.array([[25,10], [-2,1]])
y = np.array([[-1,7], [10,8]])
print(x*y) #dimensions match exactly.
```

Element-wise or Hadamard Product can be done by this:

$$(A \odot B)_{ij} = A_{ij} \, B_{ij}$$

- Basically just element multiply with element that is at the same position (as dimensions match exactly).
- **This is not matrix multiplication.**

3. What is the dot product of tensors $\mathbf{w}$ and $\mathbf{x}$

$$ \mathbf{w} =
\begin{bmatrix}
-1 & 2 & -2
\end{bmatrix}
$$

$$ \mathbf{x} =
\begin{bmatrix}
5 & 10 & 0
\end{bmatrix}
$$

```python
import torch

x = torch.tensor([-1.,2,-2])
w = torch.tensor([5.,10,0])
print(torch.dot(w,x))
```
